### Программа и ее описание

На первом этапе портируем библиотеки, читаем файл и используем базовые команды для отображения содержания датафрейма и информации о нем.

```r
library(psych)
library(ggplot2)
frame <- read.csv("~/Документы/Универ/Прога/Проект R/AppleStore.csv")
head(frame)
View(frame)
str(frame)
names(frame)
summary(frame)
sum(is.na(frame))
```

Далее сделаем несколько выборок и столбцу таблицы prime_genre

```r
frame$prime_genre == 'Games' 
frame[frame$prime_genre == 'Games', 9]
```

Следующим шагом встроенными возможностями языка с точки зрения функций математической статистики.
Сначала с помощью функции aggregate выведм в отдельную таблицу выведем среднюю цену для каждого жанра прилложения из исходного датафрейма, и переименуем названия столбцов, т.к. изначально устанавливаются стандартные названия исходя из аргументов функции.
Затем выведем описательную статистику всех столбцов датафрейма кроме тех, которые имеют тип Factor 

```r
mean_paid_genre <- aggregate(x = frame$price, by = list(frame$prime_genre), FUN = mean)
colnames(mean_paid_genre) <- c("Genre", "Mean_price") 

discr <- describeBy(x = frame[,-c(1,2,3,4,5,11,12,13)], group = list(frame$prime_genre), mat = T, digits = 3)
```

Также язык R обладает встроенными возмодностями построения графиков, но они зачастую получаются не информативными поэтому мы будем так же использовать библиотеку ggplot2.

```r
boxplot(price ~ prime_genre, frame)

ggplot(frame, aes(x = price))+
  geom_histogram(fill = "blue", col = "black", binwidth = 10)

ggplot(frame, aes(x = user_rating, fill = cont_rating))+
  geom_density(alpha = 0.3)

ggplot(frame, aes(x = user_rating, y = prime_genre, col = cont_rating))+
  geom_point(size = 5)
```

И в заключении призведем небольшую оценку датасета. На основании похожих оценок можно впоследствии обучать нейросеть и выводить предположителные ответы на заданные вопросы.
